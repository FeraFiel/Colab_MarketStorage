### Pegar tabela de .pdf

### Instalar PyPDF e pegar número total de páginas do .pdf

```python
!pip install PyPDF2
```

```python
from PyPDF2 import PdfFileReader
pdf = PdfFileReader(open(data,'rb'))
num_pages = pdf.getNumPages()
```

### Seção de import para lidar com dados e Dataframe

```python
import pandas as pd
```

# Ajuste de dados

## Append de tabelas

```python
df = pd.DataFrame() # -> cria Dataframe vazio

# -> appenda Dataframe de tabelas do .pdf enquanto número de páginas
for page in range(num_pages):
    df = df.append(tabela[page])
```

## Renomear colunas

```python
column_names = ['codigo', 'descricao', 'preco', 'unidade', 'estoque_deposito', 'estoque_gondola', 'preco_custo', 'valor_estoque']

df.set_axis(column_names, axis=1, inplace=True)
```

## Formatação de dados do Dataframe

```python
df['estoque_gondola'] = df['estoque_gondola'].apply(lambda x: x.replace('.', ''))
df['valor_estoque'] = df['valor_estoque'].apply(lambda x: x.replace('.', ''))

df['estoque_gondola'] = df['estoque_gondola'].apply(lambda x: x.replace(',', '.'))
df['valor_estoque'] = df['valor_estoque'].apply(lambda x: x.replace(',', '.'))
df['preco'] = df['preco'].apply(lambda x: x.replace(',', '.'))
df['preco_custo'] = df['preco_custo'].apply(lambda x: x.replace(',', '.'))

df['preco'] = pd.to_numeric(df['preco'])
df['preco_custo'] = pd.to_numeric(df['preco_custo'])
df['estoque_gondola'] = pd.to_numeric(df['estoque_gondola'])
df['valor_estoque'] = pd.to_numeric(df['valor_estoque'])
```

## Criar planilha para inserir dados do Dataframe

```python
nome_spreadsheet = gc.create(nome_arquivo)

nome_worksheet = nome_spreadsheet.add_worksheet(nome_arquivo, rows=len(df), cols='8')

nome_worksheet.update([df.columns.values.tolist()] + df.values.tolist())
```
